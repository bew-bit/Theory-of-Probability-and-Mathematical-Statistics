### Задача 2
О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

$D(X) = {(b-a)^2\over12}$<br/><br/>
$0.2 = {(b-0.5)^2\over12}$<br/><br/>
$b = \sqrt{0.2*12}+0.5$<br/><br/>
правая граница:

```python
b = round(m.sqrt(0.2*12)+0.5, 2)
b
```

среднее значение:

```python
(0.5+b)/2
```

### Задача 4
Рост взрослого населения города X имеет нормальное распределение.<br/>
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.<br/>
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:<br/>
а). больше 182 см<br/>
б). больше 190 см<br/>
в). от 166 см до 190 см<br/>
г). от 166 см до 182 см<br/>
д). от 158 см до 190 см<br/>
е). не выше 150 см или не ниже 190 см<br/>
ё). не выше 150 см или не ниже 198 см<br/>
ж). ниже 166 см.

```python
mean = 174
sd = 8
x = np.arange(145, 203, 0.001)
y = norm.pdf(x,mean,sd)

fig, ax = plt.subplots(figsize=(15, 10))

ax.minorticks_on()
ax.grid(which='major')
ax.grid(which='minor', linestyle=':')
ax.plot(x, y, linewidth=5)

# Вертикальные линии:
ax.vlines(174, y.min(), y.max(), color = 'g')
ax.annotate("174", xy=(173.4, -0.0015), xytext=(173.4, -0.0015), c='g')
ax.vlines(150, y.min(), y.max(), color = 'magenta')
ax.annotate("150", xy=(149.4, -0.0015), xytext=(149.4, -0.0015), c='m')
ax.vlines(158, y.min(), y.max(), color = 'magenta')
ax.annotate("158", xy=(157.4, -0.0015), xytext=(157.4, -0.0015), c='m')
ax.vlines(166, y.min(), y.max(), color = 'magenta')
ax.annotate("166", xy=(165.4, -0.0015), xytext=(165.4, -0.0015), c='m')
ax.vlines(182, y.min(), y.max(), color = 'magenta')
ax.annotate("182", xy=(181.4, -0.0015), xytext=(181.4, -0.0015), c='m')
ax.vlines(190, y.min(), y.max(), color = 'magenta')
ax.annotate("190", xy=(189.4, -0.0015), xytext=(189.4, -0.0015), c='m')
ax.vlines(198, y.min(), y.max(), color = 'magenta')
ax.annotate("198", xy=(197.4, -0.0015), xytext=(197.4, -0.0015), c='m')

# Вероятности:
ax.text(145, 0.001, '0.15%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(153, 0.0001, '2.15%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(161.1, 0.012, '13.6%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(169, 0.038, '34.1%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(174.5, 0.038, '34.1%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(182.1, 0.012, '13.6%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(190.5, 0.0001, '2.15%', fontsize='xx-large', c='steelblue', weight='heavy')
ax.text(198.5, 0.001, '0.15%', fontsize='xx-large', c='steelblue', weight='heavy')

plt.show()
```

### Задача 5
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
# Дано:
mean = 178
dispersion = 25
value = 190

# График:
x = np.arange(mean-3.5*m.sqrt(dispersion), mean+3.5*m.sqrt(dispersion), 0.001)
y = norm.pdf(x,mean,sd)

fig, ax = plt.subplots(figsize=(15, 10))

ax.minorticks_on()
ax.grid(which='major')
ax.grid(which='minor', linestyle=':')
ax.plot(x, y, linewidth=5)

# Вертикальные линии:
ax.vlines(mean, y.min(), y.max(), color = 'r')
ax.vlines(value, y.min(), y.max(), color = 'r')
ax.vlines(mean-3*sd, y.min(), y.max(), color = 'lightblue')
ax.vlines(mean-2*sd, y.min(), y.max(), color = 'lightblue')
ax.vlines(mean-sd, y.min(), y.max(), color = 'lightblue')
ax.vlines(mean+sd, y.min(), y.max(), color = 'lightblue')
ax.vlines(mean+2*sd, y.min(), y.max(), color = 'lightblue')
ax.vlines(mean+3*sd, y.min(), y.max(), color = 'lightblue')

plt.show()
```

```python
(value-mean)/m.sqrt(dispersion)
```
